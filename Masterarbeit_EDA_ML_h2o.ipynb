```python
f_data['DAMAGE'] = f_data['DAMAGE'].asfactor()  #encode the binary repsonse as a factor
f_data['DAMAGE'].levels()
```

```python
splits = f_data.split_frame(ratios=[0.7, 0.1], seed=1)  #

train = splits[0]
valid=splits[1]
test = splits[2]
```

```python
y = 'DAMAGE'
x = list(f_data.columns)
x.remove(y)
```

```python
from h2o.estimators.random_forest import H2ORandomForestEstimator
from h2o.estimators import H2OXGBoostEstimator
from h2o.estimators.gbm import H2OGradientBoostingEstimator
print('drf started')
#param = {"ntrees" : 100,"max_depth" : 5,"seed": 1, "stopping_metric":'AUC',"categorical_encoding":'one_hot_explicit'}
#rf_fit1 = H2ORandomForestEstimator(model_id='rf_fit1',ntrees=200 ,seed=1,score_each_iteration=True,stopping_metric='AUC',categorical_encoding='one_hot_explicit')#,score_each_iteration=True
#rf_fit1 = H2OXGBoostEstimator(**param,model_id='rf_fit1')#,score_each_iteration=True
rf_fit1 = H2ORandomForestEstimator(model_id='rf_fit1',ntrees=100 ,seed=1,nfolds=2,fold_assignment="Stratified",stopping_rounds=5,score_each_iteration=True,stopping_metric='AUC',categorical_encoding='one_hot_explicit')#,nfolds=2,fold_assignment='Stratified'
print('fitting started')
rf_fit1.train(x=x, y=y, training_frame=train,validation_frame=valid)#,validation_frame=valid
print('finished')
```
